```python
from scipy.stats import beta 

%matplotlib inline
```

```python
shape_par_1=0.5
shape_par_2=0.5
```

```python
import numpy as np

x_values_1 = np.arange (0.01, 1, 0.01) 

beta_func_values_1 = beta.pdf(x_values_1, shape_par_1, shape_par_2,loc=0,scale=1) 

print("Actual Mean",beta.mean(shape_par_1, shape_par_2))
print("Actual variance",beta.var(shape_par_1, shape_par_2))
```

```python
from matplotlib import pyplot as plt

plt.plot(x_values_1,beta_func_values_1)
plt.title(r'$ \alpha = 0.5 , \beta = 0.5 $',fontsize=20)

plt.savefig("./Gamma_images/dist_1.png")
```

```python
sample_1 = beta(shape_par_1, shape_par_2).rvs(size=10000)

print("mean",np.mean(sample_1))
print("variance",np.var(sample_1))

import seaborn as sns
sns.displot(sample_1, binwidth=0.01)
plt.title(r'$ \alpha = 0.5 , \beta = 0.5 $',fontsize=20)
plt.savefig("./Gamma_images/dist_sample_1.png")
```

```python
# Here Let the shape paramaters be alpha=0.5 and beta =0.5

shape_param_1 = 0.5
shape_param_2 = 0.5

# Mean of distribution : alpha/(alpha+beta)
# Variance of distribution: 

sample_means=[]

n=40

for i in range(0,1000):
    sample = beta(shape_param_1, shape_param_2).rvs(size=40)
    
    sample_mean = np.mean(sample)
    
    sample_means.append(sample_mean)
    
# print(sample_means)
    

```

```python
sns.displot(sample_means,color="g",kde="true")
plt.title(r'$ \alpha = 0.5 , \beta = 0.5 $',fontsize=20)
plt.savefig("./Gamma_images/clt.png")

print("mean",np.mean(sample_means))
print("variance",np.var(sample_means))
```

```python
print("expected mean",(shape_param_1)/(shape_param_1+shape_param_2),"....................","actual mean",np.mean(sample_means))
print("expected variance",(shape_param_1*shape_param_2)/(n*((shape_param_1+shape_param_2)**2)*(shape_param_1+shape_param_2+1)),"....................","variance",np.var(sample_means))
```
