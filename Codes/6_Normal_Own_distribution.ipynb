```python
# from scipy.stats import lognorm 

%matplotlib inline

import scipy.stats as st
import numpy
import math


def my_norm_function(x,mu,sigma):
    return (1/(sigma*np.sqrt(2*np.pi)))*np.exp(-1*0.5*np.power(((x-mu)/sigma),2,dtype=np.float64))
```

```python
import numpy as np

#function
def random_custDist(x0,x1,custDist,mu,sigma,size=None, nControl=10**6):
    samples=[]
    nLoop=0
    while len(samples)<size and nLoop<nControl:
        x=np.random.uniform(low=x0,high=x1)
        prop=custDist(x,mu,sigma)
        assert prop>=0 and prop<=1
        if np.random.uniform(low=0,high=1) <=prop:
            samples += [x]
        nLoop+=1
    return np.sort(samples)
```

```python
mu=0
sigma=1
```

```python
import numpy as np

x_values_1 = np.arange (-6, 6, 0.01)

norm_func_values_1 = my_norm_function(x_values_1,mu,sigma) 

print("Theoretical Mean:",mu)
print("Theoretical Variance:",sigma**2)
```

```python
from matplotlib import pyplot as plt

plt.plot(x_values_1,norm_func_values_1)
plt.title(r'$ \mu = 0 , \sigma = 1 $',fontsize=20)
plt.savefig("./dist_1.png")
```

```python
sample_1 = random_custDist(-6,6,my_norm_function,mu,sigma,size=10000, nControl=10**7)

print("mean",np.mean(sample_1))
print("variance",np.var(sample_1))

import seaborn as sns
sns.displot(sample_1, binwidth=0.01)
plt.xlim(-5, 5)
plt.title(r'$ \mu = 0 , \sigma = 1 $',fontsize=20)
plt.savefig("./dist_sample_1.png")
```
