```python
# from scipy.stats import lognorm 

%matplotlib inline

import scipy.stats as st
import numpy
import math
import scipy


def my_poisson_function(x,lambda_value):
    return (np.power(lambda_value,x,dtype=np.float64)*(np.exp(-lambda_value)))/(scipy.special.factorial(x))
```

```python
import numpy as np

#funtion
def random_custDist(x0,x1,custDist,lambda_value,size=None, nControl=10**7):
    samples=[]
    nLoop=0
    while len(samples)<size and nLoop<nControl:
        x=np.random.randint(low=x0,high=x1)
        prop=custDist(x,lambda_value)
        assert prop>=0 and prop<=1 , str(prop)+" "+str(x)
            
        if np.random.uniform(low=0,high=1) <=prop:
            samples += [x]
        nLoop+=1
    return np.sort(samples)
```

```python
lambda_value=1

x_values_1 = np.arange (0, 10, 1) 

poisson_values_1 = my_poisson_function(x_values_1,lambda_value) 

print("Theoretical Mean:",lambda_value)
print("Theoretical Variance:",lambda_value )

from matplotlib import pyplot as plt

plt.plot(x_values_1,poisson_values_1,marker='o', color='b')

plt.title(r'$ \lambda = 1  $',fontsize=20)
plt.savefig("./Poisson/dist_1.png")
```

```python
sample_1 = random_custDist(0,10,my_poisson_function,lambda_value,size=10000, nControl=10**7)

print("mean",np.mean(sample_1))
print("variance",np.var(sample_1))

import seaborn as sns
sns.displot(sample_1)
plt.title(r'$ \lambda = 1  $',fontsize=20)
plt.savefig("./Poisson/dist_sample_1.png")
```
