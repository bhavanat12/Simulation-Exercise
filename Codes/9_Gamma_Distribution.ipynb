```python
from scipy.stats import gamma 

%matplotlib inline
```

```python
shape_par_1=1.0 #alpha
scale_par_2=2.0 #beta
```

```python
import numpy as np

x_values_1 = np.arange (0, 10, 0.01) 

gamma_func_values_1 = gamma.pdf(x_values_1, shape_par_1,loc=0,scale=scale_par_2) 

print("Actual Mean",gamma.mean(shape_par_1, scale=scale_par_2))
print("Actual variance",gamma.var(shape_par_1, scale=scale_par_2))
```

```python
from matplotlib import pyplot as plt

plt.plot(x_values_1,gamma_func_values_1)
plt.title(r'$ \alpha = 1.0 , \theta = 2.0 $',fontsize=20)

plt.savefig("./real_gamma_distribution/dist_1.png")
```

```python
sample_1 = gamma(shape_par_1, scale=scale_par_2).rvs(size=10000)

print("mean",np.mean(sample_1))
print("variance",np.var(sample_1))

import seaborn as sns
sns.displot(sample_1, binwidth=0.01)
plt.xlim(0,10)
plt.title(r'$ \alpha = 1.0 , \theta = 2.0 $',fontsize=20)
plt.savefig("./real_gamma_distribution/dist_sample_1.png")
```

```python
# Here Let the shape paramaters be alpha=0.5 and beta =0.5

shape_param_1 = 1
scale_param_2 = 2



# Mean of distribution : alpha/(alpha+beta)
# Variance of distribution: 

sample_means=[]

n=40

for i in range(0,1000):
    sample = gamma(shape_param_1, scale=scale_param_2).rvs(size=40)
    
    sample_mean = np.mean(sample)
    
    sample_means.append(sample_mean)
    
# print(sample_means)
    

```

```python
sns.displot(sample_means,color="g",kde="true")
plt.title(r'$ \alpha = 1.0 , \theta = 2.0 $',fontsize=20)
plt.savefig("./real_gamma_distribution/clt.png")

print("mean",np.mean(sample_means))
print("variance",np.var(sample_means))
```

```python
print("expected mean",(shape_param_1*scale_param_2),"....................","actual mean",np.mean(sample_means))
print("expected variance",((shape_param_1*scale_param_2*scale_param_2)/n),"....................","variance",np.var(sample_means))
```
