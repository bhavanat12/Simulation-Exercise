```python
%matplotlib inline

import scipy.stats as st
import numpy
import math
import scipy


def my_geometric_function(x,p):
    return ((1-p)**x)*p
```

```python
import numpy as np

#funtion
def random_custDist(x0,x1,custDist,p,size=None, nControl=10**7):
    samples=[]
    nLoop=0
    while len(samples)<size and nLoop<nControl:
        x=np.random.randint(low=x0,high=x1)
        prop=custDist(x,p)
        assert prop>=0 and prop<=1 , str(prop)+" "+str(x)
            
        if np.random.uniform(low=0,high=1) <=prop:
            samples += [x]
        nLoop+=1
    return np.sort(samples)
```

```python
p=0.2

x_values_1 = np.arange(0,15,0.1)

geometric_values_1 = my_geometric_function(x_values_1,p) 

print("Theoretical Mean:",(1-p)/p)
print("Theoretical Variance:",(1-p)/(p**2) )

from matplotlib import pyplot as plt

plt.plot(x_values_1,geometric_values_1,marker='o', color='b')

plt.title(r'$ p = 0.2  $',fontsize=20)
plt.savefig("./dist_1.png")
```

```python
sample_1 = random_custDist(0,15,my_geometric_function,p,size=10000, nControl=10**7)

print("mean",np.mean(sample_1))
print("variance",np.var(sample_1))

import seaborn as sns
sns.displot(sample_1)
plt.title(r'$ p = 0.2  $',fontsize=20)
plt.savefig("./dist_sample_1.png")
```
