```python
%matplotlib inline

import scipy.stats as st
import numpy
import math
import scipy


def my_binomial_function(x, n, p):
    return scipy.special.comb(n,x)*(p**x)*((1-p)**(n-x))
```

```python
import numpy as np

#funtion
def random_custDist(x0,x1,custDist,n, p,size=None, nControl=10**7):
    samples=[]
    nLoop=0
    while len(samples)<size and nLoop<nControl:
        x=np.random.randint(low=x0,high=x1)
        prop=custDist(x,n,p)
        assert prop>=0 and prop<=1 , str(prop)+" "+str(x)
            
        if np.random.uniform(low=0,high=1) <=prop:
            samples += [x]
        nLoop+=1
    return np.sort(samples)
```

```python
n=30
p=0.2

x_values_1 = np.arange(0,25,1) 

binomial_values_1 = my_binomial_function(x_values_1,n, p) 

print("Theoretical Mean:",n*p)
print("Theoretical Variance:",n*p*(1-p) )

from matplotlib import pyplot as plt

plt.plot(x_values_1,binomial_values_1,marker='o', color='b')

plt.title(r'$ n = 30, p = 0.2  $',fontsize=20)
plt.savefig("./dist_1.png")
```

```python
sample_1 = random_custDist(0,25,my_binomial_function,n,p,size=10000, nControl=10**7)

print("mean",np.mean(sample_1))
print("variance",np.var(sample_1))

import seaborn as sns
sns.displot(sample_1)
plt.title(r'$ n = 30, p = 0.2  $',fontsize=20)
plt.savefig("./dist_sample_1.png")
```
