```python
# Import statements

%matplotlib inline

from matplotlib import pyplot as plt
import scipy.stats as st
import numpy as np
import math
import scipy
import seaborn as sns


def my_expon_function(x,lambda_value):
    temp_var = -1 * lambda_value * x
    return(lambda_value * (np.exp(temp_var, dtype=np.float64)))
```

```python
def random_custDist(x0,x1,custDist,lambda_value,size=None, nControl=10**6):
    samples=[]
    nLoop=0
    while len(samples)<size and nLoop<nControl:
        x=np.random.uniform(low=x0,high=x1)
        prop=custDist(x,lambda_value)
        assert prop>=0 and prop<=1 , str(prop)+" "+str(x)
            
        if np.random.uniform(low=0,high=1) <=prop:
            samples += [x]
        nLoop+=1
    return np.sort(samples)
```

```python
lambda_value1 = 0.5

x_values_1 = np.arange (0, 10, 0.01) 

expon_values_1 = my_expon_function(x_values_1,lambda_value1) 

print("Theoretical Mean:",1/lambda_value1)
print("Theoretical Variance:",1/(lambda_value1*lambda_value1))


plt.plot(x_values_1,expon_values_1)

plt.title(r'$ \lambda = 0.5 $',fontsize=20)
plt.savefig("./Exponential/dist_1.png")
```

```python
sample_1 = random_custDist(0,20,my_expon_function,lambda_value1,size=10000, nControl=10**7)

print("mean",np.mean(sample_1))
print("variance",np.var(sample_1))

sns.displot(sample_1, binwidth=0.1)

plt.title(r'$ \lambda = 0.5 $',fontsize=20)
plt.savefig("./Exponential/dist_sample_1.png")
```
