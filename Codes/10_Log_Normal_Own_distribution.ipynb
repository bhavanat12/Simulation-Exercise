```python
# from scipy.stats import lognorm 

%matplotlib inline

import scipy.stats as st
import numpy
import math


def my_lognorm_function(x,mu,sigma):
    return (1/x)*(1/(sigma*np.sqrt(2*np.pi)))*np.exp(-(np.log(x)-mu)**2/(2*(sigma**2)))
```

```python
import numpy as np

#funtion
def random_custDist(x0,x1,custDist,mu,sigma,size=None, nControl=10**6):
    samples=[]
    nLoop=0
    while len(samples)<size and nLoop<nControl:
        x=np.random.uniform(low=x0,high=x1)
        prop=custDist(x,mu,sigma)
        assert prop>=0 and prop<=1
        if np.random.uniform(low=0,high=1) <=prop:
            samples += [x]
        nLoop+=1
    return np.sort(samples)
```

```python
mu=0
sigma=0.5
```

```python
import numpy as np

x_values_1 = np.arange (0.01, 5, 0.01) 

log_func_values_1 = my_lognorm_function(x_values_1,mu,sigma) 

print("Theoretical Mean:",np.exp(mu+(sigma**2/2)))
print("Theoretical Variance:",(np.exp(sigma**2)-1)*np.exp(2*mu+sigma**2) )
```

```python
from matplotlib import pyplot as plt

plt.plot(x_values_1,log_func_values_1)
plt.title(r'$ \mu = 0 , \sigma = 0.5 $',fontsize=20)
plt.savefig("./Lognormal/dist_1.png")
```

```python
sample_1 = random_custDist(0,8,my_lognorm_function,mu,sigma,size=10000, nControl=10**7)

print("mean",np.mean(sample_1))
print("variance",np.var(sample_1))

import seaborn as sns
sns.displot(sample_1, binwidth=0.01)
plt.xlim(0, 5)
plt.title(r'$ \mu = 0 , \sigma = 0.5 $',fontsize=20)
plt.savefig("./Lognormal/dist_sample_1.png")
```
